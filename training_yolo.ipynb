```python
# regular boxes

def soft_nms(boxes, sigma=0.5, Nt=0.3, method=2, score_thresh=0.001):
    """
    Applies Soft Non-Maximum Suppression (Soft-NMS) to remove excessive overlapping boxes.

    Parameters:
        boxes (list of lists): List of bounding boxes in the format [x1, y1, x2, y2, confidence, class_id].
        sigma (float): Controls the suppression factor in Gaussian decay (default: 0.5).
        Nt (float): IoU threshold for suppression (default: 0.3).
        method (int):
            0 = Original NMS (hard threshold)
            1 = Linear decay
            2 = Gaussian decay (default)
        score_thresh (float): Threshold below which boxes are removed.

    Returns:
        list: Filtered boxes after applying Soft-NMS.
    """
    boxes = np.array(boxes)
    N = len(boxes)

    for i in range(N):
        max_pos = i
        max_score = boxes[i, 4]

        # Find the highest confidence box
        for j in range(i + 1, N):
            if boxes[j, 4] > max_score:
                max_score = boxes[j, 4]
                max_pos = j

        # Swap the highest confidence box to the front
        boxes[i], boxes[max_pos] = boxes[max_pos], boxes[i]

        x1_i, y1_i, x2_i, y2_i, score_i, class_id_i = boxes[i]

        # Iterate through the remaining boxes
        for j in range(i + 1, N):
            x1_j, y1_j, x2_j, y2_j, score_j, class_id_j = boxes[j]

            # Compute IoU
            xx1 = max(x1_i, x1_j)
            yy1 = max(y1_i, y1_j)
            xx2 = min(x2_i, x2_j)
            yy2 = min(y2_i, y2_j)

            w = max(0.0, xx2 - xx1 + 1)
            h = max(0.0, yy2 - yy1 + 1)
            inter = w * h
            union = ((x2_i - x1_i + 1) * (y2_i - y1_i + 1)) + ((x2_j - x1_j + 1) * (y2_j - y1_j + 1)) - inter
            IoU = inter / union if union > 0 else 0

            # Apply Soft-NMS decay based on the chosen method
            if method == 1:  # Linear
                if IoU > Nt:
                    boxes[j, 4] *= (1 - IoU)
            elif method == 2:  # Gaussian
                boxes[j, 4] *= np.exp(-(IoU**2) / sigma)
            else:  # Standard NMS
                if IoU > Nt:
                    boxes[j, 4] = 0  # Directly suppress

    # Filter out boxes with low scores
    filtered_boxes = [box.tolist() for box in boxes if box[4] > score_thresh]
    return filtered_boxes

# Example Usage:
yolo_detections = [
    [50, 50, 200, 200, 0.9, 0],  # [x1, y1, x2, y2, confidence, class_id]
    [55, 60, 195, 205, 0.85, 0],
    [200, 200, 400, 400, 0.8, 1],
    [220, 220, 420, 420, 0.75, 1]
]

filtered_detections = soft_nms(boxes, sigma=0.5, Nt=0.3, method=0)
print(filtered_detections)
print(len(filtered_detections))
```

```python
# oriented boxes

import numpy as np
import cv2
from shapely.geometry import Polygon

def rotated_iou(box1, box2):
    """
    Compute IoU for rotated bounding boxes using Shapely.

    Each box is represented as (cx, cy, w, h, angle).
    """
    # Convert (cx, cy, w, h, angle) to quadrilateral (polygon)
    def get_rotated_box(cx, cy, w, h, angle):
        angle_rad = np.radians(angle)
        cos_a = np.cos(angle_rad)
        sin_a = np.sin(angle_rad)

        # Half dimensions
        w_half = w / 2
        h_half = h / 2

        # Compute corner points relative to the center
        p1 = (cx - w_half * cos_a + h_half * sin_a, cy - w_half * sin_a - h_half * cos_a)
        p2 = (cx + w_half * cos_a + h_half * sin_a, cy + w_half * sin_a - h_half * cos_a)
        p3 = (cx + w_half * cos_a - h_half * sin_a, cy + w_half * sin_a + h_half * cos_a)
        p4 = (cx - w_half * cos_a - h_half * sin_a, cy - w_half * sin_a + h_half * cos_a)

        return Polygon([p1, p2, p3, p4])

    poly1 = get_rotated_box(*box1)
    poly2 = get_rotated_box(*box2)

    # Compute IoU
    inter_area = poly1.intersection(poly2).area
    union_area = poly1.area + poly2.area - inter_area

    return inter_area / union_area if union_area > 0 else 0

def soft_nms_obb(boxes, sigma=0.5, Nt=0.3, method=2, score_thresh=0.001):
    """
    Soft Non-Maximum Suppression for Oriented Bounding Boxes (OBB).

    boxes: list of [cx, cy, w, h, angle, confidence, class_id]
    sigma: Gaussian decay factor
    Nt: IoU threshold for suppression
    method:
        0 = Hard NMS
        1 = Linear Soft-NMS
        2 = Gaussian Soft-NMS
    score_thresh: Minimum confidence threshold to keep a box.
    """
    boxes = np.array(boxes)
    N = len(boxes)

    for i in range(N):
        max_pos = i
        max_score = boxes[i, 5]  # Confidence score

        # Find the highest confidence box
        for j in range(i + 1, N):
            if boxes[j, 5] > max_score:
                max_score = boxes[j, 5]
                max_pos = j

        # Swap the highest confidence box to the front
        boxes[i], boxes[max_pos] = boxes[max_pos], boxes[i]

        cx_i, cy_i, w_i, h_i, angle_i, score_i, class_id_i = boxes[i]

        # Iterate through remaining boxes
        for j in range(i + 1, N):
            cx_j, cy_j, w_j, h_j, angle_j, score_j, class_id_j = boxes[j]

            # Compute Rotated IoU
            IoU = rotated_iou([cx_i, cy_i, w_i, h_i, angle_i], [cx_j, cy_j, w_j, h_j, angle_j])

            # Apply Soft-NMS decay
            if method == 1:  # Linear
                if IoU > Nt:
                    boxes[j, 5] *= (1 - IoU)
            elif method == 2:  # Gaussian
                boxes[j, 5] *= np.exp(-(IoU**2) / sigma)
            else:  # Hard NMS
                if IoU > Nt:
                    boxes[j, 5] = 0  # Suppress

    # Filter out low-confidence boxes
    filtered_boxes = [box.tolist() for box in boxes if box[5] > score_thresh]
    return filtered_boxes

# Example Usage with YOLOv8-OBB Output
# obb_detections = [
#     [100, 100, 50, 30, 15, 0.9, 0],  # [cx, cy, w, h, angle, confidence, class_id]
#     [102, 105, 50, 30, 18, 0.85, 0],
#     [200, 200, 60, 40, -10, 0.8, 1],
#     [210, 210, 60, 40, -5, 0.75, 1]
# ]

filtered_obb_detections = soft_nms_obb(obb_detections, sigma=0.5, Nt=0.3, method=2)
print(filtered_obb_detections)
print(len(filtered_obb_detections))
```

```python
# regular boxes

def run_yolo_inference(image_path, score_thresh=0.2):
    """
    Runs YOLO object detection on an image and applies Soft-NMS.
    """
    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print("Error: Could not load image.")
        return

    # Run YOLO inference
    results = model(img)
    #for result in results:
    #    detections = result.boxes.data.tolist()
    #    break
    detections = results[0].cpu().boxes.data.numpy()  # Get detected boxes (x1, y1, x2, y2, confidence, class)

    # Convert YOLO output to Soft-NMS format
    yolo_boxes = [[x1, y1, x2, y2, conf, cls] for x1, y1, x2, y2, conf, cls in detections]

    # Apply Soft-NMS
    filtered_boxes = soft_nms(yolo_boxes, sigma=0.5, Nt=0.3, method=2, score_thresh=score_thresh)
    count = len(filtered_boxes)

    # Draw filtered boxes
    for x1, y1, x2, y2, conf, cls in filtered_boxes:
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        # label = f"{model.names[int(cls)]}: {conf:.2f}"
        label = str(int(cls))
        center_x = int((x1 + x2) / 2)
        center_y = int((y1 + y2) / 2)
        cv2.circle(img, (center_x, center_y), radius=5, color=(0, 255, 0), thickness=-1)  # Draw green dot
        # cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 1)
        cv2.putText(img, label, (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show result
    img = cv2.resize(img, (0,0), fx=0.75, fy=0.75)
    cv2.imshow("YOLO Detection with Soft-NMS", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imwrite(image_path[:-4] + '_processed.jpg', img)
    return(count)
```

```python
# oriented boxes

def draw_obb(image, cx, cy, w, h, angle, color=(0, 255, 255), thickness=1):
    """
    Draws an Oriented Bounding Box (OBB) on the image.
    """
    # angle_rad = np.radians(angle)
    angle_rad = angle
    cos_a = np.cos(angle_rad)
    sin_a = np.sin(angle_rad)

    w_half = w / 2
    h_half = h / 2

    p1 = (int(cx - w_half * cos_a + h_half * sin_a), int(cy - w_half * sin_a - h_half * cos_a))
    p2 = (int(cx + w_half * cos_a + h_half * sin_a), int(cy + w_half * sin_a - h_half * cos_a))
    p3 = (int(cx + w_half * cos_a - h_half * sin_a), int(cy + w_half * sin_a + h_half * cos_a))
    p4 = (int(cx - w_half * cos_a - h_half * sin_a), int(cy - w_half * sin_a + h_half * cos_a))

    cv2.line(image, p1, p2, color, thickness)
    cv2.line(image, p2, p3, color, thickness)
    cv2.line(image, p3, p4, color, thickness)
    cv2.line(image, p4, p1, color, thickness)

def run_yolo_inference(image_path, score_thresh=0.2):
    """
    Runs YOLOv8-OBB detection and applies Soft-NMS.
    """
    img = cv2.imread(image_path)
    if img is None:
        print("Error: Could not load image.")
        return

    # Run YOLOv8 inference
    results = model(img)

    # Extract Oriented Bounding Box (OBB) detections
    yolo_boxes = []
    for det in results[0].cpu().obb.data:
        cx, cy, w, h, angle, conf, cls = det[:7]  # Extract bbox details
        yolo_boxes.append([cx, cy, w, h, angle, conf, cls])

    # Apply Soft-NMS
    filtered_boxes = soft_nms_obb(yolo_boxes, sigma=0.5, Nt=0.3, method=2, score_thresh=score_thresh)
    count = np.zeros(len(model.names))

    # Draw filtered detections
    for cx, cy, w, h, angle, conf, cls in filtered_boxes:
        cx, cy, w, h = int(cx), int(cy), int(w), int(h)
        class_name = model.names[int(cls)]
        label = f"{class_name}: {conf:.2f}"
        count[int(cls)] += 1

        # Draw center dot
        if cls == 0:
            cv2.circle(img, (cx, cy), radius=5, color=(0, 255, 0), thickness=-1)
        elif cls == 1:
            cv2.circle(img, (cx, cy), radius=5, color=(241, 181, 62), thickness=-1)

        # Draw rotated bounding box (Optional)
        # draw_obb(img, cx, cy, w, h, angle)

        # Display label text
        # if cls == 0:
        #     cv2.putText(img, label, (cx - 20, cy - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        # elif cls == 1:
        #     cv2.putText(img, label, (cx - 20, cy - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 255), 1)

    # Show image
    img = cv2.resize(img, (0,0), fx=0.75, fy=0.75)
    cv2.imshow("YOLOv8-OBB with Soft-NMS", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return(count)

```
